Займемся доработками:

```python
data = data.drop(['fbs'], axis='columns')
del categorial[2]

def knn_slope_normazer(value):
    if value == 0:
        return 1.5
    return value
data['slope'] = data['slope'].apply(knn_slope_normazer)

def knn_restecg_normazer(value):
    if value > 1:
        return 0.5
    return value
data['restecg'] = data['restecg'].apply(knn_restecg_normazer)

def knn_thal_normazer(value):
    if value <= 1:
        return 2.5
    return value
data['thal'] = data['thal'].apply(knn_thal_normazer)

def knn_ca_normazer(value):
    if value > 0:
        return 1
    return 0
data['ca'] = data['ca'].apply(knn_ca_normazer)

def knn_cp_normazer(value):
    if value > 0:
        return 1
    return 0
data['cp'] = data['cp'].apply(knn_cp_normazer)
```

# L2 - регуляризация через Ridge

```python
from sklearn.linear_model import Lasso, Ridge

def ridge_reg(x_train, x_test, y_train, y_test,alpha):
    ridgereg = Ridge(alpha=alpha)
    ridgereg.fit(x_train, y_train)
    y_pred = ridgereg.predict(x_test)
    
    ret = [sum((y_pred-y_test)**2)]
    ret.extend([ridgereg.intercept_])
    ret.extend(ridgereg.coef_)
    return ret
```

```python
alpha_ridge = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20,50,100,200]
result=[]
for i in range(13):
    result.append(ridge_reg(X_train, X_test, y_train, y_test, alpha_ridge[i]))
lol1 = pd.DataFrame(result, columns=['MSE','Const'] + [x for x in X_train_scaled.columns])
lol1.index = alpha_ridge
pd.options.display.float_format = '{:,.2g}'.format
lol1
```

```python
i1 = [x for x in lol1.index if lol1['MSE'][x] == min(lol1['MSE'])][0]
```

# L1 - регуляризация через Lasso

```python
def lasso_reg(x_train, x_test, y_train, y_test,alpha):
    lassoreg = Lasso(alpha=alpha)
    lassoreg.fit(x_train, y_train)
    y_pred = lassoreg.predict(x_test)
    
    ret = [sum((y_pred-y_test)**2)]
    ret.extend([lassoreg.intercept_])
    ret.extend(lassoreg.coef_)
    return ret
```

```python
alpha_ridge = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]
result = []
for i in range(10):
    result.append(lasso_reg(X_train, X_test, y_train, y_test, alpha_ridge[i]))
lol2 = pd.DataFrame(result, columns=['MAE','Const'] + [x for x in X_train_scaled.columns])
lol2.index = alpha_ridge
pd.options.display.float_format = '{:,.2g}'.format
lol2
```
